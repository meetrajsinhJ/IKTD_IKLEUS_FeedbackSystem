```python
import os
import shutil

def combine_jpg_files(source_dir, target_dir):
    """
    Combine all JPG files from source_dir and its subdirectories into target_dir.
    """
    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    for root, dirs, files in os.walk(source_dir):
        for file in files:
            if file.lower().endswith('.jpg'):
                source_file = os.path.join(root, file)
                target_file = os.path.join(target_dir, file)

                # Check if file with same name exists in target directory
                file_exists = os.path.exists(target_file)
                i = 1
                while file_exists:
                    # If file exists, append a number to the file name
                    base, extension = os.path.splitext(file)
                    new_filename = f"{base}_{i}{extension}"
                    target_file = os.path.join(target_dir, new_filename)
                    file_exists = os.path.exists(target_file)
                    i += 1

                # Copy the file to the target directory
                shutil.copy2(source_file, target_file)
                print(f"Copied: {source_file} to {target_file}")

# Example usage
source_directory = '/Users/jadeja19/Documents/Hiwi iktd/data set/Transform dataset/Baugruppe/c'  # Replace with your source directory
target_directory = '/Users/jadeja19/Documents/Hiwi iktd/data set/Transform dataset/Baugruppe/comC'  # Replace with your target directory
combine_jpg_files(source_directory, target_directory)
```

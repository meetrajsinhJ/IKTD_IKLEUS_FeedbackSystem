```python
# Import necessary libraries
import os
from pdf2image import convert_from_path
from IPython.display import display
from PIL import Image

# Function to convert PDF to JPG
def convert_pdf_to_jpg(pdf_path, output_folder):
    """Convert each page of a PDF to a separate JPG file."""
    # Convert PDF to a list of images
    images = convert_from_path(pdf_path)

    # Create the output directory if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Save each image as a JPG file
    for i, image in enumerate(images):
        output_file = os.path.join(output_folder, f'page_{i+1}.jpg')
        image.save(output_file, 'JPEG')
        display(Image.open(output_file))

# Example usage:
# Replace 'path_to_your_pdf.pdf' with the path to your PDF file
# Replace 'path_to_output_folder' with the path to the folder where you want to save the JPGs
pdf_path = '/Users/jadeja19/Documents/Hiwi iktd/data set/solution/2019W_KL1_A3_Schrverb_Gruppenuebung_Lsg.pdf'
output_folder = '/Users/jadeja19/Documents/Hiwi iktd/data set/solution/jpg converted'
convert_pdf_to_jpg(pdf_path, output_folder)
```

```python
import os
from pdf2image import convert_from_path
from IPython.display import display
from PIL import Image

# Function to convert a single PDF to JPG
def convert_pdf_to_jpg(pdf_path, output_folder):
    images = convert_from_path(pdf_path)
    for i, image in enumerate(images):
        jpg_path = os.path.join(output_folder, f'page_{i+1}.jpg')
        image.save(jpg_path, 'JPEG')

# Function to convert all PDFs in a directory
def convert_all_pdfs(pdf_dir, output_base_dir):
    for pdf_file in os.listdir(pdf_dir):
        if pdf_file.lower().endswith('.pdf'):
            pdf_path = os.path.join(pdf_dir, pdf_file)
            output_folder = os.path.join(output_base_dir, os.path.splitext(pdf_file)[0] + '_jpg')
            os.makedirs(output_folder, exist_ok=True)
            convert_pdf_to_jpg(pdf_path, output_folder)
            print(f"Converted {pdf_file} to JPG in {output_folder}")

# Example usage
pdf_directory = '/Users/jadeja19/Documents/Hiwi iktd/data set/aufgabe1corrected'  # Replace with the directory containing your PDFs
output_base_directory = '/Users/jadeja19/Documents/Hiwi iktd/data set/Transform dataset/schraube/corrected'  # Replace with the base directory for saving output JPGs
convert_all_pdfs(pdf_directory, output_base_directory)
```
